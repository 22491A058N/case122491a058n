```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Explicitly handle quoting and escape characters
irfan = pd.read_csv('energy.csv')

# Print the dataframe to inspect its structure
print(irfan)

# Proceed with the rest of your code
srh = irfan[irfan['State'] == "Andhra"]
csk = irfan[irfan['State'] == "Karnataka"]

# Line chart
fig,diag1 = plt.subplots(2)
diag1[0].plot(srh['Date'], srh['Consumption'], color='red')
diag1[1].plot(csk['Date'], csk['Consumption'], color='blue')

# barc chart Telengana Tripura
nobii=irfan[irfan['State']=='Telengana']
cm=irfan[irfan['State']=='Tripura']

fig,diag1 = plt.subplots(2)
diag1[0].bar(nobii['Date'], nobii['Consumption'], color='red')
diag1[1].bar(cm['Date'], cm['Consumption'], color='blue')

#pie chart
State1='Delhi'
State2='Odisha'
State3='UP'
State4='Punjab'
cse5=[State1,State2]

cse4=irfan[irfan['State']==State1]['Consumption'].sum()
cse3=irfan[irfan['State']==State2]['Consumption'].sum()
cse2=[cse3,cse4]

fig,diag = plt.subplots(2)
diag[0].pie(cse2,labels=cse5,autopct='%1.1f%%')

cse5=[State3,State4]

cse4=irfan[irfan['State']==State3]['Consumption'].sum()
cse3=irfan[irfan['State']==State4]['Consumption'].sum()
cse2=[cse3,cse4]

fig,diag = plt.subplots(2)
diag[0].pie(cse2,labels=cse5,autopct='%1.1f%%')
```
